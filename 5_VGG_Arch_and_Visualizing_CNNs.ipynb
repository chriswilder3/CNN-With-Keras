# VGG Architecture :

The VGG architecture, developed by the Visual Geometry Group at the University of Oxford, is a deep convolutional neural network (CNN) primarily designed for image classification.1 It's famous for its simplicity and the significant role it played in proving that network depth is a critical component for achieving high performance in large-scale image recognition.

**The core design of VGG** models, such as VGG-16 and VGG-19, is characterized by a uniform and repetitive structure

1. **Input**: The network takes a fixed-size 4$224 \times 224$ RGB image as input.5 The only preprocessing is subtracting the mean RGB values computed over the training set.

2. **Uniform Small Filters**: The most distinctive feature is the exclusive use of very small 7$3 \times 3$ convolutional filters (with stride 1 and padding 1) throughout the entire network.8 The 9$3 \times 3$ filter is the smallest size that captures information in both the horizontal and vertical directions.

3. **Convolutional Blocks**: The convolutional layers are grouped into five blocks.11 Within each block, several 12$3 \times 3$ convolutional layers (with ReLU activation) are stacked before a max-pooling layer.

Max-Pooling: A 14$2 \times 2$ max-pooling layer with a stride of 2 is used after each convolutional block to spatially downsample the feature maps (halving the height and width).

Increasing Depth (Filters): The number of filters (or channels/depth) increases progressively after each max-pooling layer, starting typically at 64 in the first block and doubling to 128, 256, and finally 512 in the last three blocks.

Fully Connected (FC) Layers: The feature maps from the final convolutional block are flattened and passed to three fully connected layers:17The first two FC layers have 4096 channels each, followed by ReLU.

The final FC layer has 1000 channels (for the 1000 classes of the ImageNet competition) and uses a Softmax activation for classification.

# Purpose of the Architecture and Design Choices

VGG's architecture was deliberately set up the way it is to systematically investigate the effect of network depth on accuracy in large-scale image classification, a design choice motivated by the following:

1. Consistent Use of Small $3 \times 3$ FiltersThis was the key design choice, and it was superior to using a single large filter like $5 \times 5$ or $7 \times 7$ (which was common in previous networks like AlexNet) for two main reasons:

  - Increased Depth (More Non-Linearity): Stacking multiple 20$3 \times 3$ convolutional layers achieves the same receptive field (the area of the input image that a final feature map pixel sees) as a single, larger filter, but with multiple non-linear activation functions (ReLU).21 For example, three stacked 22$3 \times 3$ convolutions achieve the same 23$7 \times 7$ receptive field as a single 24$7 \times 7$ convolution.25 The three non-linear layers make the decision function more discriminative and allow the network to learn more complex features.

  - Reduced Number of Parameters: Stacking small filters is more parameter-efficient. For an input/output with 27$C$ channels, a single 28$7 \times 7$ convolution requires 29$49 C^2$ parameters, while three stacked 30$3 \times 3$ convolutions require only 31$3 \times (9 C^2) = 27 C^2$ parameters—a significant reduction.

2. Exploring Different ConvNet ConfigurationsThe VGG paper explored a series of configurations, denoted as VGG-A through VGG-E, which differed primarily in the total number of weighted layers (convolutional and fully connected layers) to test the hypothesis that deeper is better.

# Pretrained models in Keras

We can load them from keras.applications module, it has common general syntax for most pretrained models like Excetion, MobileNet, EfficientNet etc
```
keras.applications.VGG16(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
    name="vgg16",
)
```

Arguments

- include_top: whether to include the fully-connected layers at the top of the network. We can include them to use the model as it as for direct prediction,
Or we can created our own dense layers and finetune this base pretrained model.

- weights: one of None (random initialization), "imagenet" (pre-training on ImageNet), or the path to the weights file to be loaded.

- input_tensor: optional Keras tensor (i.e. output of layers.Input()) to use as image input for the model.

- input_shape: optional shape tuple, only to be specified if include_top is False (otherwise the input shape has to be (224, 224, 3) (with channels_last data format) or (3, 224, 224) (with "channels_first" data format). It should have exactly 3 input channels, and width and height should be no smaller than 32. E.g. (200, 200, 3) would be one valid value.

- pooling: Optional pooling mode for feature extraction when include_top is False.
  1. None means that the output of the model will be the 4D tensor output of the last convolutional block.
  2. avg means that global average pooling will be applied to the output of the last convolutional block, and thus the output of the model will be a 2D tensor.
  3. max means that global max pooling will be applied.

- classes: optional number of classes to classify images into, only to be specified if include_top is True, and if no weights argument is specified.

- classifier_activation: A str or callable. The activation function to use on the "top" layer. Ignored unless include_top=True. Set classifier_activation=None to return the logits of the "top" layer. When loading pretrained weights, classifier_activation can only be None or "softmax".

- name: The name of the model (string).

# Visualizing CNN Models :

Here is way to draw flowchart type diagram showing CNN layers

```
keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=False,
    show_dtype=False,
    show_layer_names=False,
    rankdir="TB",
    expand_nested=False,
    dpi=200,
    show_layer_activations=False,
    show_trainable=False,
    **kwargs
)
```

**Params :**

- model: A Keras model instance
- to_file: File name of the plot image.
- show_shapes: whether to display shape information (ie dimensions like (50,50,3)
- show_dtype: whether to display layer dtypes.
- show_layer_names: whether to display layer names.
- rankdir: rankdir argument passed to PyDot, a string specifying the format of  the plot. "TB" creates a vertical plot; "LR" creates a horizontal plot.

- expand_nested: whether to expand nested Functional models into clusters.
- dpi: Image resolution in dots per inch.
- show_layer_activations: Display layer activations (only for layers that have an activation property).
- show_trainable: whether to display if a layer is trainable

# Predict image class (Using VGG)


The pre-trained VGG16 model expects a specific input format, which you must adhere to:

1. Resize: The image must be resized to $224 \times 224$ pixels.

2. Convert to Array: Convert the image to a NumPy array.

3. Add Batch Dimension: Add a dimension to represent the "batch" of images, even if it's just one image. The required shape is $(1, 224, 224, 3)$.

4. ImageNet Preprocessing: Apply the specific pixel manipulation (like color channel shifting and conversion from RGB to BGR) that was used when the model was originally trained on ImageNet. Keras provides a convenient function for this.

````python
import PIL
PIL.Image.open("/content/image.png")
````

````python
from tensorflow.keras.utils import load_img
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
import numpy as np

image = load_img("/content/image.png", target_size=(224, 224))
  # image = PIL.Image.open("/content/image.png")

image = np.array(image)

image = np.reshape(image, (1, image.shape[0], image.shape[1], image.shape[2]))
          # or we can use
          # np.expand_dims(image, axis = 0)
          # axis: An integer or a tuple of integers specifying the position(s) in the expanded axes where the new axis (or axes) is placed.
          # A value of 0 inserts a new dimension at the beginning, -1 at the end, and so on.

image_processed = preprocess_input(image)

# Get the model's predictions (a vector of 1000 probabilities)
predictions = model.predict(image_processed)

# Decode the predictions to get the top K class labels and probabilities
# By default, it returns the top 5 predictions
decoded_predictions = decode_predictions(predictions, top=5)[0]

print("Top 5 predictions")
for i , (imagenet_id, label, score) in enumerate(decoded_predictions):
  print(f"{i+1} : {imagenet_id} : {label} : {score*100:.2f}")


# The output will list the top classes VGG16 predicts for your image, along with the confidence score (e.g., "golden_retriever: 95.34%").
````
